## Exploring Regex (Regular Expressions)

```python
import re
```

```python
#pattern , r stands for raw string
pattern = re.compile(r'abc')
```

```python
text_data = '''  
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890

Ha HaHa

MetaCharacters (Need to be escaped):
.[{()\^$|?*+

coreyms.com

321-555-4321
123.555.1234
123*555*1234

cat
bat
mat
sat

Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T'''
```

```python
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
text_data[3:6]
```

```python
# matches the url, . dot is a special charecter and need to be escaped by back slash
pattern = re.compile(r'coreyms\.com')
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
text_data[129:140]
```

```python
sentence = 'Start a scenetence and bring it to an end'
pattern = re.compile(r'^Start')
matches = pattern.finditer(sentence)
for match in matches:
    print(match)
    
pattern = re.compile(r'end$')
matches = pattern.finditer(sentence)
for match in matches:
    print(match)
```

```python
# phone numbers
pattern = re.compile(r'\d\d\d[-.]\d\d\d[-.]\d\d\d\d')
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
```

```python
# openning a file
with open('data.txt', 'r') as f:
    contents = f.read()
    print(contents)
    pattern = re.compile(r'\d\d\d[-.]\d\d\d[-.]\d\d\d\d')
    matches = pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
# more with []
with open('data.txt', 'r') as f:
    contents = f.read()
    pattern = re.compile(r'[89]00[-.]\d\d\d[-.]\d\d\d\d')
    matches = pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
# more with [x-y]
with open('data.txt', 'r') as f:
    contents = f.read()
    pattern = re.compile(r'[1-5]')
    matches = pattern.finditer(contents)
    for match in matches:
        print(match)
```

```python
# with ^
pattern = re.compile(r'[^b]at')
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
```

```python
# play with quantifiers
pattern = re.compile(r'Mr\.?\s[A-Z]\w*')
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
```

```python
# play with groups
pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text_data)
for match in matches:
    print(match)
```

```python
emails = '''
kurtjackson@bogusemail.com
Kurtjackson@university.edu
kurt-123-456-jackson@my-work.net
'''
pattern = re.compile(r'[a-zA-Z0-9.-]+@[a-zA-Z0-9.-]+')
matches = pattern.finditer(emails)
for match in matches:
    print(match)
```

```python
urls = '''
https://www.google.com
https://www.youtube.com
https://www.nasa.gov
'''
pattern = re.compile(r'https?://(www\.)?\w+\.\w+')
matches = pattern.finditer(urls)
for match in matches:
    print(match)
```

```python
#splitting into groups
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)
for match in matches:
    print(match)
```

```python
# print group 0
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)
for match in matches:
    print(match.group(0))
```

```python
# match method
sentence = 'Start a scenetence and bring it to an end'
pattern = re.compile(r'Start')
match = pattern.match(sentence)
print(match)
```

```python
# search method
sentence = 'Start a scenetence and bring it to an end'
pattern = re.compile(r'scenetence')
match = pattern.search(sentence)
print(match)
```

```python
# ignore case flag
sentence = 'Start a scenetence and bring it to an end'
pattern = re.compile(r'start', re.IGNORECASE)
match = pattern.search(sentence)
print(match)
```

```python
# ignore case flag
sentence = 'Start a scenetence and bring it to an end'
pattern = re.compile(r'start', re.I)
match = pattern.search(sentence)
print(match)
```
